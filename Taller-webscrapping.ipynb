```python
'''
Bienvenidos al taller de webscrapping con python 3 :)

soy un comentario, podrás reconocerme por ser una cadena de texto rodeado de 
triple comilla simple
'''
#Yo tambien soy un comentario, pero para una sola linea

'''
Lo primero que haremos sera incluir las librerias que necesitamos para empezar
con el webscrapping
'''
#la libreria urllib nos ayudara a recoger la informacion de internet
from urllib.request import Request, urlopen
#y beautiful soup nos ayudara a leerla
from bs4 import BeautifulSoup
```

```python
def obtener_html(url):
    '''
    Hola, yo soy la documentacion de un metodo :)
    como puedes ver soy un comentario de multiples lineas
    debajo de la cabezera del metodo
    ------
    Este metodo requiere una direccion de internet y devuelve su html
    '''
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    webpage = urlopen(req).read()
    # parsea con beautifulsoup
    html = BeautifulSoup(webpage, 'lxml')
    return html
```

```python
#Ahora podemos hacer uso de la función 'obtener_html' para 
#descaga el contenido de una web al azar
enlace = 'http://civichackers.webs.upv.es/comunidad-hc/'
html = obtener_html(enlace)
#ahora vamos a visualizar su contenido, no te asustes
print(html)
```

Como podemos ver, nuestro programa se conecta a la web deseada y descarga __TODO__ su contenido como si de un navegador se tratara. Pero esta información no es útil a menos que se trate.

```python
'''
Ejemplo sencillo, queremos extraer todos los párrafos que tiene el html
'''
parrafos = html.findAll('p')
# mostramos los elementos separados y no todo el array
for par in parrafos:
    #print('-> '+str(par))
    print('-> '+par.text)
```

```python
'''
Imaginemos que ahora quisieramos extraer los nombres de los miembros que aparecen en la página
'''
nombres = []
#con esta linea de codigo se obtiene los fragmentos html que son del tipo div y tiene un atributo data-name
fragmentos = html.findAll("div",{'data-name':True})
#vamos por cada uno de los fragmentos extrayendo su nombre
for frag in fragmentos:
    nombres.append(frag['data-name'])
print(nombres)   
#o para los valientes
#print([frag['data-name']for frag in fragmentos])
```

```python
'''
Podemos aplicar lo mismo para encontrar
los correos disponibles
'''
correos = []
fragmentos = html.findAll('a', {'href':True})
for frag in fragmentos:
    aux = frag['href']
    if 'mailto:' in aux:
        #separo la cadena por la aparicion de los :
        #y me quedo con la segunda parte
        #equivalente a borrar 'mailto:'
        aux = aux.split(':')[1]
        correos.append(aux)
print(correos)        
```

```python
'''
Vamos a mezclar la informacion para que 
sea todavia mas util
'''
#tenemos un problema, y es que todos los miembros tienen nombre pero no todos tienen correo
# la funcion len devuelve la longuitud de una estructura de datos
print(len(nombres)) #12
print(len(correos)) #10
#la solucion es 'manual', miramos a ver que miembros no tienen su correo disponible
#en este caso el quinto y el septimo
#en python los arrays empiezan los la posicion 0, por tanto la quinta posicion es el indice 4
correos.insert(4, None)
correos.insert(6, None)
#podemos crear un diccionario, como clave usamos el nombre de los usuarios y como valor su correo
midic = dict() 
for i in range(len(nombres)):
    midic[nombres[i]] = correos[i]
#Ahora puedo acceder al correo de los miembros AHC por nombre!
print(midic['Vicent Blanes'])
```

```python
'''
Ahora vamos a ver un ejemplo de como podemos guardar nuestra informacion
usaremos algo sencillo como pickel aunque podriamos montarnoslo para usar
una base de datos o una hoja de calculo
'''
#pickel es un modulo de python que permite guardar objetos en formato binario
import pickle
#ahora guardamos el diccionario
pickle.dump(midic, open('nombresycorreos.p', 'wb'))
```

```python
#vamos a cargarlo
#nos aseguramos que el objeto no exista
midic = None
midic = pickle.load(open('nombresycorreos.p', 'rb'))
#probamos...
print(midic['Vicent Blanes'])
            
```
